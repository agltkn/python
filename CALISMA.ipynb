# for döngüsü ile sayı toplatmak

```python
n = int(input("Kaç adet sayı gireceğinizi belirtiniz : "))
toplam = 0
for i in range(1, n+1):
    sayı = int(input(f"{i}. sayıyı giriniz : "))
    toplam += sayı
print("Sayılarım toplamı = {}".format(toplam))
```

# while döngüsü ile sayı toplatmak

```python
n = int(input("Kaç adet sayı gireceğinizi belirtiniz : "))
counter = 1
listem = []

while counter <= n:
    sayı = int(input(f"{counter}. sayıyı giriniz :"))
    listem.append(counter)
    counter += 1
print("Sayılarım toplamı = {}".format(sum(listem)))    
```

# map fonksiyonu

```python
sayılar = map(int, input("Toplanmasını istediğiniz sayıları giriniz : ").split())

toplam = sum(sayılar)

print("Sayılarım toplamı = {}".format(toplam)) 
```
